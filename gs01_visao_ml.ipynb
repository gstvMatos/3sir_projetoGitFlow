# Introdução

Nos últimos anos, os eventos climáticos extremos têm se tornado cada vez mais frequentes e intensos, impulsionados pelas mudanças climáticas, pelo aumento das temperaturas globais e pela degradação ambiental. Entre esses eventos, as queimadas representam um dos maiores desafios ambientais da atualidade, causando impactos significativos na biodiversidade, na qualidade do ar, na saúde pública e na economia.

No Brasil, o cenário se tornou ainda mais preocupante. Segundo dados do Instituto Nacional de Pesquisas Espaciais (INPE), o país registrou mais de **278 mil focos de queimadas em 2024**, representando um crescimento expressivo em relação aos anos anteriores. Biomas como Amazônia, Cerrado e Pantanal foram fortemente afetados, causando destruição ambiental em larga escala, aumento das emissões de carbono e prejuízos econômicos para diversos setores produtivos.

Além dos impactos ambientais, as queimadas também afetam diretamente a população. A fumaça gerada pelos incêndios compromete a qualidade do ar em diferentes regiões do país, contribuindo para o aumento de doenças respiratórias, problemas cardiovasculares e impactos na mobilidade urbana e logística.

Paralelamente, o avanço da economia espacial tem ampliado a capacidade de monitoramento do planeta através de satélites capazes de capturar dados climáticos, geográficos e ambientais em tempo real. Atualmente, organizações como o INPE e a NASA utilizam sistemas orbitais para detectar focos de calor, acompanhar eventos extremos e gerar informações estratégicas para tomada de decisão.

No entanto, grande parte das soluções existentes ainda atua de maneira reativa, identificando incêndios após o início do evento. Diante desse cenário, surge a oportunidade de utilizar Ciência de Dados e Machine Learning para transformar dados espaciais em inteligência ambiental, permitindo identificar padrões espaciais e temporais associados à ocorrência de queimadas no território brasileiro.

Com base nesse contexto, o projeto **FireWatch AI** propõe o desenvolvimento de uma solução analítica baseada em dados espaciais disponibilizados pelo INPE, com o objetivo de identificar padrões geográficos e temporais associados à ocorrência de queimadas no Brasil.

## **Hipótese 5 - A análise espacial pode auxiliar na identificação de regiões críticas**


O algoritmo KMeans foi aplicado às coordenadas geográficas (latitude e longitude) dos focos de calor para identificar agrupamentos naturais de ocorrência das queimadas no território brasileiro.

```python
from sklearn.cluster import KMeans
```

```python
X = df[['lat', 'lon']]
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
df['cluster'] = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='lon', y='lat', hue='cluster', data=df, palette='viridis', s=20, alpha=0.6)
plt.title('Clusters de Focos de Queimadas (KMeans)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
df['cluster'].value_counts()
```

```python
cluster_pct = (
    df['cluster']
    .value_counts(normalize=True)
    .sort_index()
    * 100
)

cluster_pct
```

```python
plt.figure(figsize=(8,8))

plt.pie(
    cluster_pct.values,
    labels=[f'Cluster {i}' for i in cluster_pct.index],
    autopct='%1.1f%%',
    startangle=90
)

plt.title('Distribuição Percentual dos Clusters')

plt.show()
```

Os resultados mostram que o Cluster 3 concentra aproximadamente 32,7% de todos os focos registrados na base analisada, tornando-se a principal região crítica identificada pelo modelo.

Somados, os Clusters 3 e 1 representam cerca de 53,6% dos focos observados, evidenciando uma forte concentração espacial das queimadas em determinadas regiões do país.

Esse comportamento reforça a utilidade da clusterização como ferramenta para direcionar ações de monitoramento e prevenção em áreas prioritárias.

```python
pd.crosstab(df['cluster'], df['bioma'])
```

```python
cluster_bioma = pd.crosstab(df['cluster'], df['bioma'])

cluster_bioma.plot(
    kind='bar',
    stacked=True,
    figsize=(12,6)
)

plt.title('Distribuição dos Biomas por Cluster')
plt.xlabel('Cluster')
plt.ylabel('Quantidade de Focos')
plt.show()
```

Os resultados confirmam a hipótese proposta. A aplicação do algoritmo KMeans permitiu identificar cinco agrupamentos geográficos distintos de focos de queimadas no território brasileiro.

Mesmo utilizando apenas coordenadas geográficas (latitude e longitude), os clusters apresentaram forte associação com biomas específicos, demonstrando que as queimadas seguem padrões espaciais bem definidos.

Observou-se predominância dos biomas Amazônia e Cerrado nos agrupamentos identificados, reforçando a relevância dessas regiões no cenário nacional das queimadas.

Além disso, a concentração dos focos em determinados clusters evidencia que os eventos não estão distribuídos de forma homogênea pelo país, permitindo identificar áreas prioritárias para monitoramento, fiscalização e ações preventivas.
